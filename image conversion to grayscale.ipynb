```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r"C:\Users\Pravallika\OneDrive\Documents\pawan.jpg")
```

```python
# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Display the image
plt.imshow(img_rgb)
plt.title("The Original Image")
plt.axis("off")
plt.show()
```

```python
# Print details
print("The Shape of the image is:", img_rgb.shape)
print("Top-left pixel RGB value:", img_rgb[0,0])  # instead of printing full matrix
```

```python
if img is None:
    print(" Image not found. Please check the path.")
else:
    # Convert BGR to RGB (for correct colors in matplotlib)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Convert BGR to Grayscale
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Display both images side by side
    plt.figure(figsize=(10, 5))

    # Original RGB image
    plt.subplot(1, 2, 1)
    plt.imshow(img_rgb)
    plt.title("Original RGB Image")
    plt.axis("off")

    # Grayscale image
    plt.subplot(1, 2, 2)
    plt.imshow(img_gray, cmap="gray")  # cmap="gray" shows in grayscale
    plt.title("Grayscale Image")
    plt.axis("off")

    plt.show()

 # Print details
    print(" RGB Shape:", img_rgb.shape)   # (H, W, 3)
    print(" Gray Shape:", img_gray.shape) # (H, W)  only 2D, no channels
    print("Top-left pixel RGB value:", img_rgb[0, 0])
    print("Top-left pixel Gray value:", img_gray[0, 0])
```
